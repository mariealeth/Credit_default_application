```python
import pandas as pd
from sklearn import datasets
from evidently.dashboard import Dashboard  
from evidently.report import Report
from evidently.dashboard.tabs import DataDriftTab
from evidently.metric_preset import DataDriftPreset, TargetDriftPreset, DataQualityPreset
```

```python
features_20 = pd.read_csv('features_20.csv')
```

```python
some_customers_with_nan = pd.read_csv('some_customers_with_nan.csv').iloc[:, 1:].loc[:, list(features_20['feature'])]
```

```python
trainref = pd.read_csv('X_train_with_nan.csv').iloc[:10000, 1:-1].loc[:, list(features_20['feature'])]
```

```python
import pandas as pd

from sklearn import datasets

from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

```
